```python
from langchain_community.embeddings import OllamaEmbeddings
embedding= (
    OllamaEmbeddings(model="llama3.1:8b")
    )
```

```python
r1 = embedding.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet",
        "Beta is the second letter if Greek alphabet",
    ]
)
```

```python
r1[0]
len(r1[0])
```

```python
embedding.embed_query("what is the second lettter of Greek alphabet?")
```
