#### Embedding Technique
Converting text into the vectors


```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
openai_api_key= os.getenv("OPENAI_API_KEY")
print(openai_api_key)
```

```python
from langchain_openai import OpenAIEmbeddings
embedding = OpenAIEmbeddings(model="text-embedding-3-large")
embedding
```

```python
text="We are learning the OpenAIEmbedding"
embedding_output=embedding.embed_query(text)
print(embedding_output)
```

```python
embedding_1024=OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
embedding_1024
```

```python
text="We are learning the OpenAIEmbedding"
embedding_output=embedding_1024.embed_query(text)
# print(embedding_output)
len(embedding_output)
```

```python
from langchain_community.document_loaders import TextLoader
loader=TextLoader('speech.txt')
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter= RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_document=text_splitter.split_documents(docs)
final_document
```

```python
## Vector Embedding and Vector storeDB -> ChromaDB
from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_document, embedding_1024)
db
```

```python
query="It is a distressing and oppressive duty, gentlemen of the Congress, which I have performed in thus addressing you"
reterieved_result=db.similarity_search(query)
print(reterieved_result)
```
